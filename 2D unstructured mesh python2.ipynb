```python
import numpy as np
import pandas as pd
from vtk import *
from vtk.util import numpy_support
import imp
import h5py
import datetime

from sklearn.neighbors import KDTree
import os
from tqdm import tqdm

class MeshLoad:
    
    def __init__(self, filename, mesh_type='tet'):
        
        assert(os.path.exists(filename))
        assert(mesh_type == 'tri' or mesh_type == 'tet')
        self.filename = filename
        self.mesh_type = mesh_type
        self.cells_array = None
        self.points_array = None

    def update(self):
        
        reader = vtkUnstructuredGridReader()
        reader.SetFileName(self.filename)
        reader.Update()
        
        self.data = data = reader.GetOutput()
        points = data.GetPoints()
        cells = data.GetCells()
        
        points_array = numpy_support.vtk_to_numpy(points.GetData())
        cells_array = numpy_support.vtk_to_numpy(cells.GetData())

        j = 0
        lst = []
        edges = []
        size_ = 3 if self.mesh_type == 'tri' else 4

        while j < len(cells_array):
            if  cells_array[j] == 2:
                edges.append([cells_array[j+1], cells_array[j+2]])
            elif cells_array[j] == size_:
                if size_ == 3:
                    lst.append([cells_array[j+1], cells_array[j+2], cells_array[j+3]])
                elif size_ == 4:
                    lst.append([cells_array[j+1], cells_array[j+2], cells_array[j+3], cells_array[j+4]])
            j += cells_array[j] + 1
            
        self.cells_array = np.array(lst)
        self.points_array = points_array[:,:-1] if self.mesh_type == 'tri' else points_array
        self.edges_array = np.array(edges)
        
    def write_result(self, full_x):
        array_for_write = numpy_support.numpy_to_vtk(full_x, deep=True, array_type=vtk.VTK_FLOAT)
        array_for_write.SetName("result") 
        self.data.GetPointData().AddArray(array_for_write)

        writer = vtkXMLUnstructuredGridWriter()
        writer.SetFileName('result2.vtu')
        writer.SetInputData(self.data)

        writer.Write()
```

```python
time_start = datetime.datetime.now()
ml = MeshLoad('./very_large_mesh.vtk', mesh_type='tri')
ml.update()
```

```python
# Must be computed once for each mesh

elements_to_reduction = []
top = []
bottom = []
for n in xrange(len(ml.edges_array)):
    e1 = ml.edges_array[n][0]
    e2 = ml.edges_array[n][1]
    if ml.points_array[e1][1] == 0 and ml.points_array[e2][1] == 0:
        if e1 not in elements_to_reduction:
            elements_to_reduction.append(e1)
        if e2 not in elements_to_reduction:
            elements_to_reduction.append(e2)
        if e1 not in bottom:
            bottom.append(e1)
        if e2 not in bottom:
            bottom.append(e2)
    if ml.points_array[e1][1] == 1 and ml.points_array[e2][1] == 1:
        if e1 not in elements_to_reduction:
            elements_to_reduction.append(e1)
        if e2 not in elements_to_reduction:
            elements_to_reduction.append(e2)
        if e1 not in top:
            top.append(e1)
        if e2 not in top:
            top.append(e2)
```

```python
# Number of points
NN = len(ml.points_array)

#Number of elements
NE = len(ml.cells_array)

# Const
A = 1
```

```python
# Global stiffness matrix
K_glob = np.zeros((NN, NN))
```

```python
ds_dxi = [-1, 1, 0]
ds_deta = [-1, 0, 1]

def integral(i, j, J_e):
    J_inv = np.linalg.inv(J_e) # Inverse of the Jacobian
    m1 = ds_dxi[j]*J_inv[0][0] + ds_deta[j]*J_inv[0][1]
    m2 = ds_dxi[i]*J_inv[0][0] + ds_deta[i]*J_inv[0][1]
    m3 = ds_dxi[j]*J_inv[1][0] + ds_deta[j]*J_inv[1][1]
    m4 = ds_dxi[i]*J_inv[1][0] + ds_deta[i]*J_inv[1][1]
    return A * (m1*m2 + m3*m4) * np.linalg.det(J_e) * 0.5
```

```python
# Derivatives of the shape functions for triangular elements
ds = np.array([
    [-1, 1, 0],
    [-1, 0, 1]
])

k = 0
for e in ml.cells_array:
    coordinates = np.array([
    [ml.points_array[e[0]][0], ml.points_array[e[0]][1]],
    [ml.points_array[e[1]][0], ml.points_array[e[1]][1]],
    [ml.points_array[e[2]][0], ml.points_array[e[2]][1]]
    ])
    
    # Jacobian of the element
    J_e = np.dot(ds , coordinates)
    
    # Element stiffness matrix
    K_e = np.zeros((3, 3))
    for i in xrange(3):
        for j in xrange(3):
            K_e[i][j] = integral(i, j, J_e)

    for i in xrange(3):
        for j in xrange(3):
            p = ml.cells_array[k][i]
            q = ml.cells_array[k][j]            
            K_glob[p][q] += K_e[i][j]
    k += 1
```

```python
# Force vector
F = np.zeros((NN, 1))
```

```python
# Boundary vector
Q = np.zeros((NN, 1))

for v in bottom:
    column = K_glob[:,v]
    for i in xrange(NN):
        Q[i] -= column[i] * 1 # EBC value
        
K_reduction = np.delete(K_glob, elements_to_reduction, axis=1)
K_reduction = np.delete(K_reduction, elements_to_reduction, axis=0)
Q_reduction = np.delete(Q, elements_to_reduction, axis=0)
```

```python
T = np.zeros((NN, 1))
for e in bottom:
    T[e] = 1

res = np.linalg.solve(K_reduction, Q_reduction)

n = 0
for k in xrange(NN):
    if k not in elements_to_reduction:
        T[k] = res[n]
        n += 1       


ml.write_result(T)
time_end = datetime.datetime.now()

print(time_end - time_start)
```
